```python
# Import data
nz_df = pd.read_csv("../datasets/nz_indicators.csv", encoding = "latin")
```

```python
new_df = nz_df[['year', 'LIH_B60_contemp_median_AHC', 'LIH_B50_contemp_median_AHC', 'LIH_B50_CV_AHC', 'LIH_B60_CV_AHC',
               'LIH_B40_contemp_median_AHC', 'LIEH_B60_contemp_median_AHC', 'LIEH_B50_contemp_median_AHC', 
               'LIEH_B50_CV_median_AHC', 'LIEH_B60_CV_median_AHC', 'LIS65H_B60_CV_median_AHC', 'LISPH_B60_CV_median_AHC',
               'LIS65H_B50_CV_median_AHC', 'LISPH_B50_CV_median_AHC', 'child_poverty_ahc_perc_HH_B50_CV',
               'child_poverty_ahc_perc_HH_B60_CV', 'child_poverty_ahc_perc_HH_B40_contemp_median',
               'child_poverty_ahc_perc_HH_B50_contemp_median', 'child_poverty_ahc_perc_HH_B60_contemp_median',
               'child_poverty_ahc_perc_HH_B60_CV_one_or_more_FT_working_parents',
               'child_poverty_ahc_perc_HH_B60_CV_one_or_more_PT_working_parents',
                'teriary_education_participation_rate', 'education_spend_as_percent_of_GDP', 
                'education_spend_as_percent_of_total_gov_expenses', 'median_teriary_loan_as_perc_of_med_inc_AHC',
                'median_teriary_loan_balance_as_perc_of_med_inc_AHC', 'university_affordability_ratio_to_mean_salary',
                'polytechnics_affordability_ratio_to_mean_salary', 'wananga_affordability_ratio_to_mean_salary', 
                'bachelors_earning_premium_hourly', 'dips_cert_earning_premium_hourly', 'school_earning_premium_hourly',
                'bachelors_earning_premium_weekly', 'dips_cert_earning_premium_weekly', 'school_earning_premium_weekly'
               ]]
new_df.head()
```

```python
# Function that removes day, month and time from year column
def remove_chars(x):
    x = x.replace(x[-5:], '')
    return x.replace(x[:6], '')
```

```python
pd.options.mode.chained_assignment = None
new_df['year'] = new_df['year'].apply(remove_chars)
```

```python
new_df = new_df.set_index('year')
new_df = new_df.sort_index()
new_df
```

```python
new_df.fillna(method='ffill')
```

```python
poverty_df = new_df[['LIH_B60_contemp_median_AHC']]
poverty_df.rename(columns={"LIH_B60_contemp_median_AHC": "poverty"}, inplace=True)
poverty_df = poverty_df.drop(poverty_df[poverty_df['poverty'] == '-'].index)
poverty_df.astype(int).plot()
```

```python
kid_poverty_df = new_df[['child_poverty_ahc_perc_HH_B60_contemp_median']]
kid_poverty_df.rename(columns={"child_poverty_ahc_perc_HH_B60_contemp_median": "poverty"}, inplace=True)
kid_poverty_df = kid_poverty_df.drop(kid_poverty_df[kid_poverty_df.poverty == 'None'].index)
kid_poverty_df.astype(int).plot()
```

```python
def remove_last(x):
    return x.replace(x[-2:], '')
```

```python
edu_num_df = new_df[['teriary_education_participation_rate']]
edu_num_df.rename(columns={"teriary_education_participation_rate": "rate"}, inplace=True)
edu_num_df = edu_num_df.dropna(how='all')
edu_num_df['rate'] = edu_num_df['rate'].apply(remove_last)
edu_num_df.head()
```

```python
edu_num_df.astype(float).plot()
```

```python
poverty_edu_df = new_df[['LIH_B60_contemp_median_AHC', 'teriary_education_participation_rate']][new_df.index.astype(int) > 1993]
poverty_edu_df = poverty_edu_df.rename(columns={"teriary_education_participation_rate": "rate", "LIH_B60_contemp_median_AHC": "poverty"})
poverty_edu_df = poverty_edu_df.drop(poverty_edu_df[poverty_edu_df.poverty == '-'].index)
poverty_edu_df['rate'] = poverty_edu_df['rate'].apply(remove_last)
poverty_edu_df.head()
```

```python
corr = poverty_edu_df['poverty'].astype('float64').corr(poverty_edu_df['rate'].astype('float64'))
corr
```

```python
poverty_edu_df.astype('float').plot.scatter(x='poverty', y='rate')
```
